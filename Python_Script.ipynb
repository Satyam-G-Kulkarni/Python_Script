```python
import requests

def fetch_all_data(api_url):
    all_data = []
    page = 1
    while True:
        try:
            response = requests.get(api_url, params={'page': page})
            if response.status_code == 200:
                data = response.json()
                all_data.extend(data['data']['data'])
                if data['data']['current_page'] == data['data']['last_page']:
                    break  # Break if we've fetched all pages
                else:
                    page += 1  # Move to the next page
            else:
                print("Error:", response.status_code)
                break
        except Exception as e:
            print("An error occurred:", e)
            break
    return all_data

def group_citations_by_response(data):
    grouped_citations = {}
    for message in data:
        if 'source' in message:
            sources = message['source']
            response = message.get('response')
            for source in sources:
                source_id = source.get('id')
                source_link = source.get('link')
                if response and source_link:
                    if response not in grouped_citations:
                        grouped_citations[response] = []
                    grouped_citations[response].append({"ID": source_id, "Link": source_link})
        else:
            print("No source found for message ID:", message.get('id'))
    return grouped_citations

def fetch_and_group_citations(api_url):
    all_data = fetch_all_data(api_url)
    grouped_citations = group_citations_by_response(all_data)
    return grouped_citations

def main():
    api_url = "https://devapi.beyondchats.com/api/get_message_with_sources"
    grouped_citations = fetch_and_group_citations(api_url)

    # Display citations in a formatted text output
    for response, citations in grouped_citations.items():
        print('citations =')
        for citation in citations:
            print(f"  - ID: {citation['ID']}")
            print(f"    Link: {citation['Link']}")
        print()

if __name__ == "__main__":
    main()
```
